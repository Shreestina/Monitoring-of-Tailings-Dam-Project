```python
import json
# load data using Python JSON module
with open('data/sample.json','r') as f:
    data = json.loads(f.read())
# Flatten data
df_samples1_list = pd.json_normalize(data, record_path =['otu_ids'], meta=['Node_id'])
```

```python
df_samples1_list
```

```python
df_samples2_list = pd.json_normalize(data, record_path =['sample_values'], meta=['Node_id'])
df_samples2_list
```

```python
merged_samples = pd.concat([df_samples1_list,df_samples2_list],axis=1)
merged_samples
```

```python
merged_samples.columns=["sample_date","Node_id","sample_values","Node_id(2)"]
merged_samples.drop(['Node_id(2)'], axis=1, inplace=True)
merged_samples
```

```python
first_column=merged_samples.pop('Node_id')
merged_samples.insert(0,'Node_id',first_column)
merged_samples
```

```python
merged_samples.to_csv('data/sample.csv')
```
